```python
def word_frequency(text):
 
    words = text.lower().split()
   
    word_count = {}
    
    for word in words:
        
        word = word.strip('.,!?":;()[]')
        
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    
    return word_count

# Example usage
text = "Hello world! Welcome to the world of Python. Python is great."
frequency = word_frequency(text)

for word, count in frequency.items():
    print(f"'{word}': {count}")
```

```python
class InventorySystem:
    def __init__(self):
        
        self.inventory = {}

    def add_item(self, item_name, quantity, price):
        # Add or update an item in the inventory
        if item_name in self.inventory:
            self.inventory[item_name]['quantity'] += quantity
        else:
            self.inventory[item_name] = {'quantity': quantity, 'price': price}
        print(f"Added {quantity} of '{item_name}' to the inventory.")

    def update_item(self, item_name, quantity=None, price=None):

        if item_name in self.inventory:
            if quantity is not None:
                self.inventory[item_name]['quantity'] = quantity
            if price is not None:
                self.inventory[item_name]['price'] = price
            print(f"Updated item '{item_name}'.")
        else:
            print(f"Item '{item_name}' not found in inventory.")

    def remove_item(self, item_name):
       
        if item_name in self.inventory:
            del self.inventory[item_name]
            print(f"Removed item '{item_name}' from inventory.")
        else:
            print(f"Item '{item_name}' not found in inventory.")

    def view_inventory(self):
        
        if not self.inventory:
            print("Inventory is empty.")
        else:
            print("Inventory:")
            for item_name, details in self.inventory.items():
                print(f"{item_name} - Quantity: {details['quantity']}, Price: ${details['price']}")

inventory = InventorySystem()

inventory.add_item('Apple', 50, 0.5)
inventory.add_item('Banana', 30, 0.2)
inventory.add_item('Orange', 40, 0.6)

inventory.view_inventory()

inventory.update_item('Apple', quantity=60, price=0.55)

inventory.remove_item('Banana')

inventory.view_inventory()
```

```python
class InventoryManager:
    def __init__(self):
        self.inventory = {}

    def add_product(self, name, quantity, price):
        if name in self.inventory:
            print(f"Product '{name}' already exists. Use update to modify the quantity or price.")
        else:
            self.inventory[name] = {'quantity': quantity, 'price': price}
            print(f"Product '{name}' added to inventory.")

    def update_quantity(self, name, quantity):
        if name in self.inventory:
            self.inventory[name]['quantity'] += quantity
            print(f"Product '{name}' quantity updated. New quantity: {self.inventory[name]['quantity']}")
        else:
            print(f"Product '{name}' not found in inventory.")

    def calculate_total_value(self):
        total_value = sum(item['quantity'] * item['price'] for item in self.inventory.values())
        print(f"Total inventory value: ${total_value:.2f}")
        return total_value

    def low_stock_report(self, threshold):
        low_stock_items = {name: details for name, details in self.inventory.items() if details['quantity'] < threshold}
        if low_stock_items:
            print("Low-stock items:")
            for name, details in low_stock_items.items():
                print(f"{name}: {details['quantity']} units (Price: ${details['price']:.2f})")
        else:
            print("No low-stock items found.")

inventory = InventoryManager()

inventory.add_product("Laptop", 50, 1000)
inventory.add_product("Smartphone", 30, 600)
inventory.add_product("Headphones", 100, 150)

inventory.update_quantity("Laptop", 20)

inventory.calculate_total_value()

inventory.low_stock_report(40)
```

```python
gradebook = {
    "Ali": {"Math": 85, "Science": 90, "History": 78},
    "Babar": {"Math": 92, "Science": 88, "History": 75},
    "Faisal": {"Math": 78, "Science": 85, "History": 82},
    "Dania": {"Math": 95, "Science": 93, "History": 89}
}

def calculate_average(grades):
    return sum(grades.values()) / len(grades)

average_grades = {student: calculate_average(grades) for student, grades in gradebook.items()}

def find_highest_lowest(subject):
    highest_grade = None
    lowest_grade = None
    highest_student = ""
    lowest_student = ""
    
    for student, grades in gradebook.items():
        if subject in grades:
            grade = grades[subject]
            if highest_grade is None or grade > highest_grade:
                highest_grade = grade
                highest_student = student
            if lowest_grade is None or grade < lowest_grade:
                lowest_grade = grade
                lowest_student = student
                
    return (highest_student, highest_grade), (lowest_student, lowest_grade)

def sort_students_by_average():
    return sorted(average_grades.items(), key=lambda x: x[1], reverse=True)

print("Average Grades for each student:")
for student, avg in average_grades.items():
    print(f"{student}: {avg:.2f}")

highest_math, lowest_math = find_highest_lowest("Math")
print(f"\nHighest grade in Math: {highest_math[0]} with grade {highest_math[1]}")
print(f"Lowest grade in Math: {lowest_math[0]} with grade {lowest_math[1]}")

sorted_students = sort_students_by_average()
print("\nStudents sorted by average grade:")
for student, avg in sorted_students:
    print(f"{student}: {avg:.2f}")
```

```python
class ECommerceStore:
    def __init__(self):
        self.products = {
            "apple": {"name": "Apple", "price": 1.2, "quantity": 100},
            "banana": {"name": "Banana", "price": 0.5, "quantity": 150},
            "orange": {"name": "Orange", "price": 1.0, "quantity": 120},
            "grape": {"name": "Grape", "price": 2.5, "quantity": 80},
        }
        self.cart = {}
        self.discount_rate = 0
        self.tax_rate = 0.1

    def add_to_cart(self, product_name, quantity):
        if product_name in self.products and self.products[product_name]["quantity"] >= quantity:
            if product_name in self.cart:
                self.cart[product_name]["quantity"] += quantity
            else:
                self.cart[product_name] = {
                    "name": self.products[product_name]["name"],
                    "price": self.products[product_name]["price"],
                    "quantity": quantity
                }
            self.products[product_name]["quantity"] -= quantity
            print(f"{quantity} {product_name}(s) added to cart.")
        else:
            print(f"Sorry, not enough stock for {product_name} or product not available.")

    def remove_from_cart(self, product_name, quantity):
        if product_name in self.cart:
            if self.cart[product_name]["quantity"] >= quantity:
                self.cart[product_name]["quantity"] -= quantity
                self.products[product_name]["quantity"] += quantity
                print(f"{quantity} {product_name}(s) removed from cart.")
                if self.cart[product_name]["quantity"] == 0:
                    del self.cart[product_name]
            else:
                print(f"Not enough {product_name} in cart to remove.")
        else:
            print(f"{product_name} is not in the cart.")

    def calculate_total(self):
        total = 0
        for item in self.cart.values():
            total += item["price"] * item["quantity"]
        return total

    def apply_discount(self, discount_percentage):
        self.discount_rate = discount_percentage

    def apply_tax(self, total):
        return total * (1 + self.tax_rate)

    def checkout(self):
        total = self.calculate_total()
        total_after_discount = total * (1 - self.discount_rate / 100)
        total_after_tax = self.apply_tax(total_after_discount)

        print("\n--- Checkout Summary ---")
        print(f"Total before discount: ${total:.2f}")
        print(f"Discount applied: {self.discount_rate}%")
        print(f"Total after discount: ${total_after_discount:.2f}")
        print(f"Tax applied: {self.tax_rate * 100}%")
        print(f"Total after tax: ${total_after_tax:.2f}")

store = ECommerceStore()

store.add_to_cart("apple", 5)
store.add_to_cart("banana", 3)

store.apply_discount(10)
store.checkout()

store.remove_from_cart("apple", 2)

store.checkout()
```
